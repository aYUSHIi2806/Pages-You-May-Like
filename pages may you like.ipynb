```python
import json
```

```python
def upload_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)
    return data
```

```python
data=upload_data("data.json")
print(data)
```

```python
def display_udata(data):
    print("user and there connection")
    for user in data["users"]:
        print(f" {user['id']}: {user['name']} is a friend of {user['friends']} and like page is {user['liked_pages']}")
    print("pages")
    for page in data["pages"]:
        print(f"page id is  : {page['id']} and name of page is = {page['name']}")

data=display_udata(data)
```
